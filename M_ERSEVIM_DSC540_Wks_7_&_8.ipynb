```python
metdf["Object Begin Date"] = metdf["Object Begin Date"].astype(str).astype(int) #TRANSFORM 5 - make sure Date is INT and binnable
```

```python
metdf['binned'] = pd.cut(x=metdf['Object Begin Date'], bins=[-9999, -2000, -1000, 0, 1000, 1200, 1400, 1600, 1800, 2000, 2100])
pd.value_counts(metdf['binned']) #TRANSFORM 6 - define year bins above, then count how many in each
```

```python
table2 = pd.pivot_table(data=metdf,index=['Country','binned']) #TRANSFORM 7 - pivot looking for strange coountry/date combos
table2 # for example, United States with an object date at least 3600 years before the U.S. was founded, lol
```
