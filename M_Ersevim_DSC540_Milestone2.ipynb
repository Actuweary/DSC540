```python
import pandas as pd #importing Pandas to handle dataframe
```

```python
df = pd.read_excel('SUB-IP-EST2019-ANNRESb.xlsx', engine='openpyxl') # needed openpyxl to handle ".xlsx" format...
```

```python
# STEP 1
cols_to_keep = ['Geographic Area','Census']
df2 = df.loc[:, cols_to_keep] #just keep the first two columns
```

```python
df2 #check
```

```python
#STEP 2
df3 = df2.dropna() #delete the footer and any rows with no numbers in the census column
```

```python
df3
```

```python
#STEP 3
df4 = df3.rename(columns={'Geographic Area': 'City/State', 'Census': 'Count'}) #rename the columns
```

```python
df4
```

```python
# STEP 4
df4['City/State'] = df4['City/State'].str.replace('city,', '') #remove most of the common descriptions leaving 
df4['City/State'] = df4['City/State'].str.replace('town,', '') # just the city and state
df4['City/State'] = df4['City/State'].str.replace('village,', '')
df4['City/State'] = df4['City/State'].str.replace('borough,', '')
```
